# ## Exercise 1 — Clean Sensor Data for ML
#
# You have raw sensor readings.
#
# Invalid values:
#
# - negative numbers
# - `9999`
#
# Tasks:
#
# - a=create a clean list with only valid readings
# - b=count raw readings
# - c=count valid readings
# - d=count invalid readings
# - e=calculate average valid reading
# - f=find minimum valid reading
# - g=find maximum valid reading
# - h=print a clean report
#
# Goal:
#
# Practice filtering, list comprehension, numeric operations, and reporting.

````python
readings = [22.5, 23.1, -5.0, 24.0, 9999, 22.8, 21.9, -1.0, 23.7]

a=[i for i in readings if i>=0 and i !=9999]
print(f"a={a}")


invalid=[i for i in readings if i<0 or i==9999 ]
print(f"invalid={invalid}")

b=len(readings)
print(f"b={b}")

c=len(a)
print(f"c={c}")

d=len(invalid)
print(f"d={d}")

e=sum(a)/len(a)
print(f"e={e}")

f=min(a)
print(f"f={f}")

g=max(a)
print(f"g={g}")
````

# ## Exercise 2 — Normalize Feature Values
#
# You have numeric feature values for a simple ML model.
#
# Normalization rule:
#
# ```python
# normalized_value = value / max_value
# ```
#
# Tasks:
#
# - find the maximum value
# - create a normalized list
# - print original values
# - print normalized values
# - print each normalized value with 3 decimal places
#
# Goal:
#
# Practice feature scaling logic before NumPy and Pandas.


````python
feature_values = [120, 300, 250, 400, 150, 500]

max_value=max(feature_values)
normalized_list=[i/max_value for i in feature_values]

print(f"max_value={max_value}")
print(f"normalized_list={normalized_list}")

decimal_3=[f"{i:.3f}" for i in normalized_list]
decimal_3
````

# ## Exercise 4 — Clean Text Labels
#
# You have messy class labels from a dataset.
#
# Tasks:
#
# - remove extra spaces
# - convert all labels to lowercase
# - remove empty labels
# - remove duplicate labels while keeping order
# - print the final cleaned label list
#
# Goal:
#
# Practice text cleaning for dataset labels.

````python
raw_labels = [" Cat ", "dog", "DOG ", " car", "", "cat", " Person ", "person", "  "]

remove_spaces=[i.strip() for i in raw_labels ]
lowercase=[i.lower() for i in remove_spaces]
remove_empty=[i for i in lowercase if not i==""]
remove_duplicate=set(remove_empty)

print(f"final: {remove_duplicate}")
````

# ## Exercise 5 — Train/Test Split Manually
#
# You have a small dataset represented as a list.
#
# Tasks:
#
# - split the data manually into 80% train and 20% test
# - do not use any library
# - calculate the split index
# - create `train_data`
# - create `test_data`
# - print both lists
# - print their lengths
#
# Goal:
#

````python
data = [12, 15, 18, 21, 24, 30, 33, 36, 40, 45]

len_data=len(data)
print(f"len_data: {len_data}")

split_data=int(0.8*len_data)
print(f" split_data: {split_data}")

train_data=data[:8]
print(f"train_data: {train_data}")

test_data=data[8:]
print(f"test_data: {test_data}")


print(f"len_train_data: {len(train_data)}")

print(f"len_test_data: {len(test_data)}")
````
