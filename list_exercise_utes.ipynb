# ## Exercise 1 — Clean Sensor Data for ML
#
# You have raw sensor readings.
#
# Invalid values:
#
# - negative numbers
# - `9999`
#
# Tasks:
#
# - a=create a clean list with only valid readings
# - b=count raw readings
# - c=count valid readings
# - d=count invalid readings
# - e=calculate average valid reading
# - f=find minimum valid reading
# - g=find maximum valid reading
# - h=print a clean report
#
# Goal:
#
# Practice filtering, list comprehension, numeric operations, and reporting.

````python
readings = [22.5, 23.1, -5.0, 24.0, 9999, 22.8, 21.9, -1.0, 23.7]

a=[i for i in readings if i>=0 and i !=9999]
print(f"a={a}")


invalid=[i for i in readings if i<0 or i==9999 ]
print(f"invalid={invalid}")

b=len(readings)
print(f"b={b}")

c=len(a)
print(f"c={c}")

d=len(invalid)
print(f"d={d}")

e=sum(a)/len(a)
print(f"e={e}")

f=min(a)
print(f"f={f}")

g=max(a)
print(f"g={g}")
````

````python
values = [10, 12, 11, 13, 500, 14, 9, 15, 600, 10]

average=sum(values)/len(values)
normal_values=[]
for i in values:
    if not i >2*average:
        normal_values.append(i)


print(normal_values)


````
