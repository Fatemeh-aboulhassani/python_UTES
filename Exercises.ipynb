d = {"a":10, "b":5, "c":20}
Keep only values greater than 10

```python
d = {"a":10, "b":5, "c":20}
n={k:v for k,v in d.items() if v>10}
n
```

d = {"name":"ali", "city":"tehran"}
Convert all keys to uppercase

```python
d = {"name":"ali", "city":"tehran"}
n={k.upper():v for k,v in d.items()}
n
```

d = {"a":10, "b":5, "c":20}
If value > 10 → multiply it by 2

```python
d = {"a":10, "b":5, "c":20}
n={k:2*v for k,v in d.items() if v>10}
n
```
