```python
#Count the frequency of each character in the string
#text = "banana"
#{'b':1, 'a':3, 'n':2}

text = "banana"

count = {}

for a in text:
    if a in count:
        count[a] += 1
    else:
        count[a] = 1

print(count)

```

```python
text="fatemeh aboulhassani chimeh"

count={}

for char in text:
    if char in count:
        count[char]+=1
    else: 
        count[char]=1


print(count)
```
